#### Average Daily Returns
We can see below that the daily returns for both stocks follow a normal distribution. In simple terms, this shows that the returns around the mean occur more frequently than returns further from the mean. However, real-life data rarely ever follows a *perfectly* normal distribution. We can therefore use the skewness and kurtosis coefficients to measure how different the respective distributions are compared to the normal distribution.

```python
df_AMD['Daily Return'] = df_AMD['Adj Close'].pct_change()
sns.displot(data=df_AMD['Daily Return'].dropna(), bins=400, kde=True)
plt.title('Distribution of Daily Returns - AMD')
plt.xlabel('Daily Return (%)')
plt.show()
```

```python
df_NVDA['Daily Return'] = df_NVDA['Adj Close'].pct_change()
sns.displot(df_NVDA['Daily Return'].dropna(), bins=320, kde=True, color='tab:blue')
plt.title('Distribution of Daily Returns - NVDA')
plt.xlabel('Daily Return (%)')
plt.show()
```

In the below calculations, I will use Fisher's kurtosis which subtracts '3' from the Pearson's kurtosis in order to make it easier to compare with the normal distribution (a normal dist. would have zero kurtosis under Fisher's definition). Also note the use of bias=False to calculate the *sample* skewness and kurtosis (not the population statistic).

```python
a = stats.kurtosis(df_AMD, axis=0, fisher=True, bias=False)[5]
b = stats.skew(df_AMD, axis=0, bias=False)[5]
c = stats.kurtosis(df_NVDA, axis=0, fisher=True, bias=False)[5]
d = stats.skew(df_NVDA, axis=0, bias=False)[5]

print('------ AMD -------',
      f'Excess Kurtosis: {a:.2f}',
      f'Skewness: {b:.2f}',
      ' ',
      '------ NVDA -------',
      f'Excess Kurtosis: {c:.2f}',
      f'Skewness: {d:.2f}',
      sep='\n'
      
)
```

Excess kurtosis of both distributions are greater than 3 (with NVDA being significantly more so than AMD) suggesting that both are **leptokurtic** distributions. That is, both tend to produce more outliers than the normal dist. In a financial context, leptokurtic distributions indicate high risk as you are more likely to encounter extreme values (of returns).

The skewness values are both positive indicating that the tail lies on the right side of the distribution (so in this context, we can say that AMD and NVDA tend to produce more positive returns than negative).

### 3. Building the ARIMA model
Before we begin forecasting, an important thing to note in time series (TS) analysis is the notion of stationarity. Stationarity is a crucial assumption in TS analysis as it implies that our variables have statistical properties (i.e. means, variances, autocorrelations, etc) that are constant over time. Since forecasting fundamentally relies on historical relationships being a reliable guide for the future, we require this assumptions so that we can extrapolate results. In this project, I will use the Augmented Dickey-Fuller test (ADF) to test for non-stationarity.

The null and alternative hypotheses for the ADF test are:

$H_0$: a unit root is present, thus the series is non-stationary

$H_1$: the series is stationary


```python
result = adfuller(df_AMD['AMD'])
result2 = adfuller(df_NVDA['NVDA'])
print('------ AMD -------',
      f'ADF Test Statistic: {result[0]:.2f}',
      f'5% Critical Value: {result[4]["5%"]:.2f}',
      f'p-value: {result[1]:.2f}',
      ' ',
      '------ NVDA -------',
      f'ADF Test Statistic: {result2[0]:.2f}',
      f'5% Critical Value: {result2[4]["5%"]:.2f}',
      f'p-value: {result2[1]:.2f}',
      sep='\n'
)
```

As seen above, the p-values for both tests are much greater than 0.05 (our chosen significance level) so we fail to reject the null hypothesis of non-stationarity. Thus, as expected, both time series are non-stationary. Next, we'll address this issue and transform them into stationary series by (first) differencing. That is:

### $$ y'_{t} = y_{t} - y_{t-1} $$

```python
df_AMD['Difference'] = df_AMD['AMD'].diff()
df_NVDA['Difference'] = df_NVDA['NVDA'].diff()

diff1 = adfuller(df_AMD['Difference'].dropna())
diff2 = adfuller(df_NVDA['Difference'].dropna())

print('------ AMD -------',
      f'ADF Test Statistic: {diff1[0]:.2f}',
      f'5% Critical Value: {diff1[4]["5%"]:.2f}',
      f'p-value: {diff1[1]:.2f}',
      ' ',
      '------ NVDA -------',
      f'ADF Test Statistic: {diff2[0]:.2f}',
      f'5% Critical Value: {diff2[4]["5%"]:.2f}',
      f'p-value: {diff2[1]:.2f}',
      sep='\n'
)
```

After the transformation, we can now reject the null hypothesis of non-stationarity (as the p-value < 0.05). Since we've satisfied the assumption of stationarity, we can move onto forecasting. For this project, I have chosen to use an ARIMA model to forecast the stock prices. The general ARIMA formula is as follows

### $$  y'_{t} = c + \phi_{1} y'_{t-1} + ... + \phi_{p} y'_{t-p} + \theta_{1} \epsilon_{t-1} + ...\theta_{q} \epsilon_{t-q} + \epsilon_{t}  $$

where $p$ is the order of the autoregressive (AR) term and $q$ is the order of the moving average (MA) term. We can think of $p$ as the number of lags to be used as predictors, and $q$ as the number of lagged forecast errors.


#### Choosing *p*:
We can inspect the Partial Autocorrelation (PACF) plot to determine the number of required AR-terms:

```python
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
```

```python
# AMD model
diff_AMD = df_AMD['Difference'].dropna()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,5))

ax1.plot(diff_AMD)
ax1.set_title('First Difference (AMD)')
ax2.set_ylim(0, 1)
plot_pacf(diff_AMD, ax=ax2);                        # semicolon req. otherwise plots twice (pacf bug)
```

We can observe from the PACF plot that the 2nd lag is above the significance line, thus we'll choose $ p_{AMD} = 2 $

```python
# NVDA model
diff_NVDA = df_NVDA['Difference'].dropna()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,5))

ax1.plot(diff_NVDA)
ax1.set_title('First Difference (NVDA)')
ax2.set_ylim(0, 1)
plot_pacf(diff_NVDA, ax=ax2);                        # semicolon req. otherwise plots twice (pacf bug)
```

We can observe from the PACF plot that the 7th lag is above the significance line, thus we'll choose $ p_{NVDA} = 7 $

#### Choosing *q*:
We can inspect the ACF plot to determine the number of required MA-terms:

```python
# AMD model
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,5))

ax1.plot(diff_AMD)
ax1.set_title('First Difference (AMD)')
ax2.set_ylim(0, 1)
plot_acf(diff_AMD, ax=ax2);    
```

We can observe from the ACF plot that the 2nd lag is above the significance line, thus we'll choose $ q_{AMD} = 2 $

```python
# NVDA model
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,5))

ax1.plot(diff_NVDA)
ax1.set_title('First Difference (NVDA)')
ax2.set_ylim(0, 1)
plot_acf(diff_NVDA, ax=ax2);    
```

We can observe from the ACF plot that the 7th lag is above the significance line, thus we'll choose $ q_{NVDA} = 7 $.

#### Fitting the ARIMA model:

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
# drop NaN vals
df_AMD.dropna(inplace=True)
df_NVDA.dropna(inplace=True)
# turn index to datetime w/ Daily freq
df_AMD.index = pd.DatetimeIndex(df_AMD.index).to_period('D')
df_NVDA.index = pd.DatetimeIndex(df_NVDA.index).to_period('D')
```

```python
# AMD model
model = ARIMA(df_AMD['AMD'], order=(2, 1, 2))        # where order = (p, d, q), where d = difference
result = model.fit(disp=0)
print(result.summary())
```

// p-values of our AR and MA terms are << 0.05 indicating high significance.

```python
# copy resids but w/ no datetime freq as residual plot doesn't take freq
result_nd = result.resid.copy()
result_nd.index = result_nd.index.to_timestamp(freq=None)
```

```python
# plot residual errors
resid_AMD = pd.DataFrame(result.resid)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,5))
ax1.plot(result_nd)
ax1.title.set_text('Residuals (AMD)')
ax2.hist(resid_AMD, density=True)
ax2.title.set_text('Density')
```

Plotting the residuals allows us to ensure that there are no patterns (we want constant mean and variance). Although we clearly have constant mean around zero, the variance has a slight increasing trend.

```python
# NVDA model (takes a while)
model_2 = ARIMA(df_NVDA['NVDA'], order=(7, 1, 7))        # where order = (p, d, q), where d = difference
result_2 = model_2.fit(disp=0)
print(result_2.summary())
```

// most p-values of our AR and MA terms are << 0.05 indicating high significance.

```python
# copy resids but w/ no datetime freq as residual plot doesn't take freq
result2_nd = result_2.resid.copy()
result2_nd.index = result2_nd.index.to_timestamp(freq=None)
```

```python
# plot residual errors
resid_NVDA = pd.DataFrame(result_2.resid)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,5))
ax1.title.set_text('Residuals (NVDA)')
ax1.plot(result2_nd)
ax2.hist(resid_NVDA, density=True)
ax2.title.set_text('Density')
```

We can observe that we have a uniform variance, and the residuals have a mean around zero

```python
resid_NVDA.mean()
```

```python
# Actual vs Fitted values (AMD)
result.plot_predict(
    start=500,
    end=560,
    dynamic=False
)
# Actual vs Fitted values (NVDA)
result_2.plot_predict(
    start=500,
    end=560,
    dynamic=False
)
plt.show()
```

Note that when we set dynamic=False the *in-sample* lagged values are used for prediction. So this can make the fitted forecast and actuals look artificially good (from the graphs above we can see the forecasts are very close to the actual values).

Thus, we can say that these are decent ARIMA models though we cannot say that these are the *best* ARIMA models. For that, we would need to do an out-of-sample forecasting.

###  4. Out of Sample Forecasting

```python
# clean up dataframes
df_AMD.drop(columns=['Difference'], inplace=True)
df_NVDA.drop(columns=['Difference'], inplace=True)
```

#### Train-Test split

```python
# we'll take 10% as testing
# note len(df_AMD) = len(df_NVDA)
n = int(len(df_AMD) * 0.9)     

train_AMD = df_AMD['AMD'][:n]
test_AMD = df_AMD['AMD'][n:]

train_NVDA = df_NVDA['NVDA'][:n]
test_NVDA = df_NVDA['NVDA'][n:]
```

```python
print(len(train_AMD))
print(len(test_AMD))
```

#### Build Arima models on the training data

```python
# training set AMD model (had to drop one MA and AR term as they were highly insignificant)
model_3 = ARIMA(train_AMD, order=(1, 1, 1))        
result_3 = model_3.fit(disp=0)
print(result_3.summary())
```

```python
# error warned me of non-stationarity in NVDA terms so I'm rechecking the training data
ad = adfuller(train_NVDA)
print(
      '------ NVDA -------',
      f'ADF Test Statistic: {ad[0]:.2f}',
      f'5% Critical Value: {ad[4]["5%"]:.2f}',
      f'p-value: {ad[1]:.2f}',
      sep='\n'
)
```

p-value clearly suggests a rejection of the null-hypothesis (we fail to reject non-stationarity). May need to second difference.

```python
# turn series to dataframe
train_NVDA = train_NVDA.to_frame()

# difference again
train_NVDA['Difference'] = train_NVDA['NVDA'].diff()
```

```python
# check for non-stationarity
ad2 = adfuller(train_NVDA['Difference'].dropna())

print(
      '------ NVDA -------',
      f'ADF Test Statistic: {ad2[0]:.2f}',
      f'5% Critical Value: {ad2[4]["5%"]:.2f}',
      f'p-value: {ad2[1]:.2f}',
      sep='\n'
)
```

p-value is now < 0.05. We had to difference the training data for NVDA one more time to remove non-stationarity.

```python
# drop NaN
diff_NVDA_2 = train_NVDA['Difference'].dropna()

diff_NVDA_ = diff_NVDA_2.copy()
diff_NVDA_.index = diff_NVDA_2.index.to_timestamp(freq=None)
```

```python
# NVDA model choosing p

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,5))

ax1.plot(diff_NVDA_)
ax1.set_title('Second Difference (NVDA)')
ax2.set_ylim(0, 1)
plot_pacf(diff_NVDA_2, ax=ax2);                        # semicolon req. otherwise plots twice (pacf bug)
```

```python
# NVDA model choosing q

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,5))

ax1.plot(diff_NVDA_)
ax1.set_title('Second Difference (NVDA)')
ax2.set_ylim(0, 1)
plot_acf(diff_NVDA_2, ax=ax2);  
```

```python
# training set NVDA model
model_4 = ARIMA(train_NVDA['NVDA'], order=(3, 1, 2))   
result_4 = model_4.fit(disp=0)
print(result_4.summary())
```

```python
# we'll do a 5 day-ahead forecast
step = 5

fc_AMD, se_AMD, conf_AMD = result_3.forecast(step)
```

```python
fc_NVDA, se_NVDA, conf_NVDA = result_4.forecast(step)
```

```python
# turn into pandas series
fc_AMD = pd.Series(fc_AMD, index=test_AMD[:step].index)
lower_AMD = pd.Series(conf_AMD[:, 0], index=test_AMD[:step].index)
upper_AMD = pd.Series(conf_AMD[:, 1], index=test_AMD[:step].index)
```

```python
# remove freq from index
test_AMD.index = test_AMD.index.to_timestamp(freq=None)
test_NVDA.index = test_NVDA.index.to_timestamp(freq=None)
```

```python
# plot AMD forecast
plt.plot(test_AMD[:step], label='actual')
plt.plot(fc_AMD, label='forecast')
plt.fill_between(lower_AMD.index, lower_AMD, upper_AMD, color='k', alpha=0.1)
plt.title('Actual vs Forecast')
plt.legend(loc='upper left')
```

```python
fc_NVDA = pd.Series(fc_NVDA, index=test_NVDA[:step].index)
lower_NVDA = pd.Series(conf_NVDA[:, 0], index=test_NVDA[:step].index)
upper_NVDA = pd.Series(conf_NVDA[:, 1], index=test_NVDA[:step].index)
```

```python
# plot NVDA forecast
plt.plot(test_NVDA[:step], label='actual')
plt.plot(fc_NVDA, label='forecast')
plt.fill_between(lower_NVDA.index, lower_NVDA, upper_NVDA, color='k', alpha=0.1)
plt.title('Actual vs Forecast')
plt.legend(loc='upper left')
```

### Conclusion:

Whilst this project taught me a lot about ARIMA modelling and how to implement it in python, the results clearly show that stock price prediction using ARIMA models cannot be succesful. Why? The answer simply lies in its definition. The ARIMA model is based on the idea that information hidden in *past values* of a time series can **alone** predict future values; so we can see how such a definition could not be compatible with the stock market (a highly volatile environment with plenty of factors affecting future prices). Not to mention, many (including Nobel prize winner Eugene Fama) have argued that stock price movements are simply impossible to predict in the short-term. In the future, I hope to explore this area of stock price prediction further by potentially looking into Support Vector Machines and Neural Networks. I also believe that it would prove more beneficial for me to focus more on predicting trend *direction* rather than short-term price movement.
